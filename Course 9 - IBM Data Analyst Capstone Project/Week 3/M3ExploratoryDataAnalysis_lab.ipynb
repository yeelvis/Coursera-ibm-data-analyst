## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python

df['ConvertedComp'].plot(kind='box', figsize=(8, 6))

plt.title('Distribution Data of Converted Compensation')
plt.ylabel('Annual USD Salary')

plt.show()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
Q1 = df['ConvertedComp'].quantile(0.25)
Q3 = df['ConvertedComp'].quantile(0.75)
IQR = Q3 - Q1
IQR
```

Find out the upper and lower bounds.


```python
# your code goes here
upper_bounds = Q3 + 1.5 * IQR
print('Upper Bounds :', upper_bounds)
lower_bounds = Q1 - 1.5*IQR
print('Lower Bounds :', lower_bounds)
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
outlier =[]

df_outlier = df[(df['ConvertedComp'] > upper_bounds) | (df['ConvertedComp'] < lower_bounds)]
df_outlier.shape
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
df_remove_outlier = df[(df['ConvertedComp'] <= upper_bounds) & (df['ConvertedComp'] >= lower_bounds)]
```

## Correlation


### Finding correlation


Find the correlation between `Age` and all other numerical columns.
